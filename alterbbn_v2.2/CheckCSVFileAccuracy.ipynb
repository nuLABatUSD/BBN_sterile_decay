## a's

```python
df_a_np = pd.read_csv("CSV Files/a_np.csv", names=["index", "output"], index_col=False, usecols = ["output"])
df_a_np_hannah = pd.read_csv("300-7e-05-FullTestNew/mass_300_mix_7e-05_a_np.csv", names=["index", "output"], index_col=False, usecols = ["output"])

fig = plt.figure()

for frame in [df_a_np, df_a_np_hannah]:
    plt.plot(frame.index, frame['output'])
    
sum_np_diff = sum(df_a_np_hannah['output'] - df_a_np['output'])
print(sum_np_diff / sum(df_a_np['output'])) 

plt.show()
```

```python
df_a_rhonu = pd.read_csv("CSV Files/a_rhonu.csv", names=["index", "output"], index_col=False, usecols = ["output"])
df_a_rhonu_hannah = pd.read_csv("300-7e-05-FullTestNew/mass_300_mix_7e-05_a_rhonu.csv", names=["index", "output"], index_col=False, usecols = ["output"])

fig = plt.figure()

for frame in [df_a_rhonu, df_a_rhonu_hannah]:
    plt.plot(frame.index, frame['output'])
    #plt.show()
    
sum_rhonu_diff = sum(df_a_rhonu_hannah['output'] - df_a_rhonu['output'])
print(sum(df_a_rhonu_hannah['output'] - df_a_rhonu['output']))

plt.show()
```

```python
df_a_dqdt = pd.read_csv("CSV Files/a_dqdt.csv", names=["index", "output"], index_col=False, usecols = ["output"])
df_a_dqdt_hannah = pd.read_csv("300-7e-05-FullTestNew/mass_300_mix_7e-05_a_dQdt.csv", names=["index", "output"], index_col=False, usecols = ["output"])

fig = plt.figure()

for frame in [df_a_dqdt, df_a_dqdt_hannah]:
    plt.plot(frame.index, frame['output'])
    
sum_dqdt_diff = sum(df_a_dqdt_hannah['output'].head(253) - df_a_dqdt['output'].head(253))
print(sum_dqdt_diff)

plt.show()
```
